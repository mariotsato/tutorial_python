```python
def capital_indexes(name):
    list = []
    n = len(name)
    for i in range(n):
        if name[i].isupper() == True:
            list += [i]
    return list
           
capital_indexes('MAriO')
```

```python
def mid1(name):
    n = len(name)
    m = n / 2
    i1 = int(m)
    
    if n % 2 == 0:
        return ''
    else:
        return name[i1]
        
mid1('Renata')
```

```python
statuses = {
    "Giovanna" : "online",
    "Rebeca" : "online",
    "Lais" : "online",
    "Luis" : "offline"
}


def online_count(dictionary):
    count = 0
    k = "online"
    for i in dictionary:
        if dictionary[i] == k:
            count += len(dictionary[i])
    return count

online_count(statuses)
```

```python
def only_ints(x, y):
    if type(x) is int and type(y) is int:
        answer = True
    else:
        answer = False
    return answer

only_ints("a", 2)
```

```python
def double_letters(lett):
    let = lett.replace(" ", "")
    n = len(let) - 1
    answer = []
    
    for i in range(n):
        if str(let[i]) == str(let[i + 1]):
            answer.append(True)
        else:
            answer.append(False)
    print(answer)

    if True in answer:
        return True
    else:
        return False

double_letters("lets do that")
```

```python
def add_dots(text):
    answer = ""
    for i in text:
        answer = answer + i + "."
        new_answer = answer[:-1]
    return new_answer

add_dots("congratulations")


def remove_dots(text):
    answer = text.replace(".", "")
    return answer

remove_dots("a.a.a.a.a.g.a.j.a.n.f.h.f.k.a")
```

```python
def count(word):
    lst = []
    c = "-"
    for pos,char in enumerate(word):
        if(char == c):
            lst.append(pos)

    n = len(lst) + 1
    return n


count("ho-tel")
```

```python
def is_anagram(a, b):
    list_a = []
    list_b = []
    for i in a: # i is String
        list_a.append(i)
        list_a.sort()
        print(list_a)
    for i in b: 
        list_b.append(i)
        list_b.sort()
        print(list_b)
    if list_a == list_b:
        return True
    else:
        return False

is_anagram("tese", "test")
```

```python
def flatten(lst):
    lst_final = []
    for i in lst:
        for j in i:
            lst_final.append(j)
    return lst_final

flatten([[1, 2], [3, 4]])
```

```python
def largest_difference(lst):
    for i in lst:
        if i == max(lst):
            maxim = i 
        elif i == min(lst):
            minim = i
    sub = maxim - minim
    return sub

largest_difference([-5,2,4,9,10])
            
```

```python
def get_row_col(text):
    dict_r = {"A1": (0, 0),
              "A2": (1, 0), 
              "A3": (2, 0), 
              "B1": (0, 1), 
              "B2": (1, 1), 
              "B3": (2, 1), 
              "C1": (0, 2),
              "C2": (1, 2),
              "C3": (2, 2)
             }

    if text in dict_r:
        return dict_r[text]

get_row_col("A3")
```

```python
def palindrome(word):
    if word[::-1] == word:
        return True
    else:
        return False

palindrome("abbbbba")
```

```python
def up_down(numb):
    tuple1 = ()
    down = numb - 1
    up = numb + 1
    tuple1 = (down, up)
    return tuple1

up_down(6)
```
